```python
def dict_invert(d):
    '''
    d: dict
    Returns an inverted dictionary according to the instructions above
    
    If d = {1:10, 2:20, 3:30} then dict_invert(d) returns {10: [1], 20: [2], 30: [3]}
    If d = {4:True, 2:True, 0:True} then dict_invert(d) returns {True: [0, 2, 4]}
    '''
    output = {} #output variable is going to be a dictionary

    for i in d.keys(): #loop through "d" variable with "i" as the index variable.
        if d[i] in output: #if the value of d[i] is already in the output variable
            output[d[i]].append(i) #add "i" to the output key. d[i] is always a list, so the method "append" works
        else:
            output[d[i]] = [i] #Add the key d[i] it the output and add [i] as the values
    for val in output.values(): #loop through the output
        val.sort() #Sort the values wtihin the list keys

    return output

dict_invert({1:10, 2:20, 3:30})
dict_invert({4:True, 2:True, 0:True})
```

```python
def laceStringsRecur(s1, s2):
    """
    s1 and s2 are strings.

    Returns a new str with elements of s1 and s2 interlaced,
    beginning with s1. If strings are not of same length, 
    then the extra elements should appear at the end.
    """
    def helpLaceStrings(s1, s2, out):
        if s1 == '':
            return out + s2
        if s2 == '':
            return out + s1
        else:
            return helpLaceStrings(s1[1:], s2[1:], out + s1[0] + s2[0])
    return helpLaceStrings(s1, s2, '')

laceStringsRecur("abc", "123")
```
